```python
import pandas as pd

data = pd.read_csv('DATA FILE.csv')
data.head()
```

```python
import numpy as np

# Replace * with NaN in entire dataframe
data = data.replace("*", np.nan)

data.head()
```

```python
data.columns
```

```python
data['Date'].head()
```

```python
data['Date'] = pd.to_datetime(data['Date'], format="%b-%y")

data = data.sort_values("Date")
data.set_index("Date", inplace=True)

data.head()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data
```

```python
import numpy as np

cols_with_invalid_zero = ["NSE", "GP", "CPI", "ER", "BCR"]

data[cols_with_invalid_zero] = data[cols_with_invalid_zero].replace(0, np.nan)

data = data.dropna()
```

```python
data
```

```python
data.columns
```

```python
cols = ["NSE", "GP", "CPI", "ER", "BCR", "IIP"]

for c in cols:
    data[c] = data[c].astype(str).str.replace(",", "", regex=False)
    data[c] = pd.to_numeric(data[c], errors="coerce")

data = data.dropna()
```

```python
data
```

```python
data.dtypes
```

```python
import statsmodels.api as sm

y = data["NSE"]
X = data[["GP", "CPI", "ER", "BCR", "IIP"]]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
plt.plot(data.index, data["NSE"], label="NSE")
plt.plot(data.index, data["GP"], label="Gold Price (GP)")
plt.plot(data.index, data["CPI"], label="CPI")
plt.plot(data.index, data["ER"], label="Exchange Rate (ER)")
plt.plot(data.index, data["BCR"], label="Brent Crude Oil (BCR)")
plt.plot(data.index, data["IIP"], label="IIP")

plt.title("Macroeconomic Variables Over Time")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.grid()
plt.show()
```
